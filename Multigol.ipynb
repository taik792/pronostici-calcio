```python
+11.88
1# Sostituisci TUONOME e NOMEPROGETTO con i tuoi dati reali di GitHub
!wget import numpy as np
from scipy.stats import poisson

def ai_oracle_v21_3_talking():
    print("AI ORACLE v21.3 - STRATEGY & VALUE ANALYST")
    print("------------------------------------------")

    try:
        # --- INPUT DATI ---
        q1 = float(input("Quota 1: "))
        qx = float(input("Quota X: "))
        q2 = float(input("Quota 2: "))
        qgg = float(input("Quota Goal: "))
        qng = float(input("Quota No Goal: "))

        print("\n[ GOL SQUADRA CASA ]")
        c_o05_q = float(input("Quota Over 0.5 Casa: "))
        c_o15_q = float(input("Quota Over 1.5 Casa: "))
        print("\n[ GOL SQUADRA OSPITE ]")
        o_o05_q = float(input("Quota Over 0.5 Ospite: "))
        o_o15_q = float(input("Quota Over 1.5 Ospite: "))

        # --- MOTORE DI CALCOLO ---
        l_casa = -np.log(max(0.01, 1 - (1/c_o05_q)))
        l_ospite = -np.log(max(0.01, 1 - (1/o_o05_q)))
        if (1/c_o15_q) < 0.45: l_casa *= 0.85
        if (1/o_o15_q) < 0.45: l_ospite *= 0.85

        # --- CALCOLO PROBABILITÀ PER LE 3 SCELTE ---
        def calc_p(mi, ma, segno):
            p = 0
            for c in range(8):
                for o in range(8):
                    cond_segno = (c>=o if segno=="1X" else o>=c)
                    if cond_segno and mi<=(c+o)<=ma:
                        p += poisson.pmf(c, l_casa) * poisson.pmf(o, l_ospite)
            return p * 100

        segno_fav = "1X" if q1 <= q2 else "X2"
        p_15 = calc_p(1, 5, segno_fav)
        p_25 = calc_p(2, 5, segno_fav)
        p_35 = calc_p(3, 5, segno_fav)

        # --- RICHIESTA QUOTE PER CONFRONTO ---
        print(f"\n--- INSERISCI LE QUOTE DEL TUO SITO PER {segno_fav} ---")
        bk_15 = float(input(f"Quota {segno_fav} + MG 1-5: "))
        bk_25 = float(input(f"Quota {segno_fav} + MG 2-5: "))
        bk_35 = float(input(f"Quota {segno_fav} + MG 3-5: "))

        # --- ANALISI STRATEGICA ---
        results = [
            (p_15, bk_15, "MG 1-5", "COPERTURA TOTALE (Include 1-0)"),
            (p_25, bk_25, "MG 2-5", "MODERATA (Perdi su 1-0 / 0-1)"),
            (p_35, bk_35, "MG 3-5", "AGGRESSIVA (Solo goleade)")
        ]

        print("\n" + "="*50)
        print(f"🎯 REPORT STRATEGICO PER {segno_fav}")
        print("="*50)

        for prob, quota, nome, nota in results:
            q_reale = 100 / prob
            diff = ((quota / q_reale) - 1) * 100

            print(f"\n▶️ {segno_fav} + {nome} | Prob: {prob:.1f}%")
            print(f"   Quota Bookie: {quota:.2f} | Quota Reale: {q_reale:.2f}")

            if diff > 5:
                print(f"   ✅ VALORE: +{diff:.1f}% - OTTIMA GIOCATA!")
            elif diff > -3:
                print(f"   🟡 EQUA: {diff:.1f}% - Accettabile in multipla.")
            else:
                print(f"   ❌ NO VALUE: {diff:.1f}% - Stai regalando soldi al bookmaker.")
            print(f"   💡 NOTA: {nota}")

        print("\n" + "="*50)

    except Exception as e:
        print(f"Errore: {e}")

if __name__ == "__main__":
    ai_oracle_v21_3_talking()
```
